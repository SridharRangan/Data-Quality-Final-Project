```python
plt.figure(figsize=(16,8)) #Creating figure size and background in Seaborn
sns.set_style("whitegrid")
plt.title('Grouping players by Age', fontsize=20, fontweight='bold', y=1.05,) # title of graph
plt.xlabel('Number of players', fontsize=15) # x-axis and y-axis labels and fonts
plt.ylabel('Players Age', fontsize=15)
sns.countplot(x="Age", data=data1, palette="Paired"); #specifiying age, dataset and color scheme
plt.show() #displaying graph
```

```python
plt.figure(figsize=(16,8)) #Figure size and style for seaborn
sns.set_style("whitegrid")
plt.title('Grouping players by Potential', fontsize=20, fontweight='bold', y=1.05,) # title and x label and y label for axis
plt.xlabel('Number of players', fontsize=15)
plt.ylabel('Players Age', fontsize=15)
sns.countplot(x="Potential", data=data1, palette="Paired"); #specifying count plot with potential, dataset and color scheme
plt.show() # displaying results
```

```python
facetexp = data1[data1['Club'].isin(['FC Barcelona', 'Real Madrid CF', 'FC Bayern Munich', 'Juventus', 'Inter'])] #experimenting with facets
a = sns.FacetGrid(facetexp, col="Club") #facet grid by club
a.map(sns.kdeplot, "Overall") #visualizing distribution of scores on each team
```

```python
sns.pairplot(data1[['Overall', 'Potential', 'Aggression']]) #pair plot experimentation with overall, potential and aggression
```

```python
data1_p = data1.groupby(['Age'])['Potential'].mean() #new dataframes for potential and overall
data1_o = data1.groupby(['Age'])['Overall'].mean()

data1_summary = pd.concat([data1_p, data1_o], axis=1) # concating both data frames into summary

ax = data1_summary.plot()
ax.set_ylabel('Rating')
ax.set_title('Average Rating by Age')
```
